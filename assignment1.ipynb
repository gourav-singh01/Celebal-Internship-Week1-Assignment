```python
import pandas as pd
import numpy as np

df = pd.read_csv("Combined_dataset.csv")

print(df.shape)
print(df.dtypes)
df.head()
```

```python
df['product_description'] = df['product_description'].fillna('No Description Available')
df['seller_name'] = df['seller_name'].fillna('Unknown Seller')

df['rating'] = df['rating'].fillna(df['rating'].median())
df['ratings_count'] = df['ratings_count'].fillna(0)

if df.duplicated().sum() > 0:
    df = df.drop_duplicates()
```

```python
if df['initial_price'].dtype == 'object':
    df['initial_price'] = df['initial_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)

df['initial_price'] = pd.to_numeric(df['initial_price'], errors='coerce').fillna(0)

np.random.seed(42)
df['quantity'] = np.random.randint(1, 6, size=len(df))

df['total_amount'] = df['initial_price'] * df['quantity']

high_rated_products = df[df['rating'] > 4.0]
cols_to_show = ['product_id', 'title', 'initial_price', 'quantity', 'total_amount', 'rating']
high_rated_products[cols_to_show].head()
```

```python
df.to_csv("Cleaned_Combined_dataset.csv", index=False)
```
